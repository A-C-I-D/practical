```python
length_width = ['sepal.length' , 'sepal.width' , 'petal.length' , 'petal.width']
```

```python
for col in length_width:
    print(f"for {col} variance is {df[col].var()}")
```

```python
for col in length_width:
    print(f"for {col} max is {df[col].max()}")
```

```python
for col in length_width:
    print(f"for {col} min is {df[col].min()}")
```

```python
for col in length_width:
    print(f"for {col} std deviation is {df[col].std()}")
```

```python
for col in length_width:
    print(f"for {col} max is {df[col].quantile(0.25)}")
```
